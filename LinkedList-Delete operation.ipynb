```python
#delete operation by value with the insertion at end on linked list
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class LinkedList:
    def __init__(self):
        self.head=None
    def iae(self,data):
        newnode=Node(data)
        if not self.head:
            self.head=newnode
            return
        current=self.head
        while current.next:
            current=current.next
        current.next=newnode
    def deletevalue(self,key):
        current=self.head
        if not current:
            print("Empty linked list")
            return
        if current.data==key:
            self.head=current.next
            print(f"{key} deleted from list")
            return
        prev=None
        while current and current.data!=key:
            prev=current
            current=current.next
        if not current:
            print(f"{key} not found")
            return
        prev.next=current.next
        print(f"{key} deleted from LL")
    def display(self):
        current=self.head
        if not current:
            print("LL-Empty")
            return 
        while current:
            print(current.data, end="---")
            current=current.next
        print("None")
ll=LinkedList()
while True:
    print("\n LinkedList-Insert...")
    print("1.Insert")
    print("2.Display")
    print("3.Delete By Value")
    print("4.Exit")
    choice=input("Enter your choice:")
    if choice=='1':
        data=int(input("Enter a value to insert:"))
        ll.iae(data) 
    elif choice=='2':
        ll.display()
    elif choice=='3':
        key=int(input("Enter the value to delete:"))
        ll.deletevalue(key)
    elif choice=='4':
        print("Exit")
        break
    else:
        print("Enter the valid choice(1/2/3/4):")
```

```python
#delete operation at end  with the insertion at end on linked list
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class LinkedList:
    def __init__(self):
        self.head=None
    def iae(self,data):
        newnode=Node(data)
        if not self.head:
            self.head=newnode
            return
        current=self.head
        while current.next:
            current=current.next
        current.next=newnode
    def deleteend(self):
        if self.head is None:
            print("Empty-LL")
            return
        if self.head.next is None:
            print(f"Deleted the last node {self.head.data}")
            self.head=None
            return
        current=self.head
        while current.next.next:
            current=current.next
        print(f" Deleted last node{current.next.data}")
        current.next=None
    def display(self):
        current=self.head
        if not current:
            print("LL-Empty")
            return 
        while current:
            print(current.data, end="---")
            current=current.next
        print("None")
ll=LinkedList()
while True:
    print("\n LinkedList-Insert...")
    print("1.Insert")
    print("2.Display")
    print("3.Delete at end")
    print("4.Exit")
    choice=input("Enter your choice:")
    if choice=='1':
        data=int(input("Enter a value to insert:"))
        ll.iae(data) 
    elif choice=='2':
        ll.display()
    elif choice=='3':
        ll.deleteend()
    elif choice=='4':
        print("Exit")
        break
    else:
        print("Enter the valid choice(1/2/3/4):")
```
